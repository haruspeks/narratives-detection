```python
# @title Entity linking (disambiguate)
from enum import Enum
import wikipedia

import nltk
from nltk.corpus import stopwords

from unidecode import unidecode
import re

# Create a set of stop words
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))

# Define a function to remove stop words from a sentence
def remove_stop_words(sentence):
  # Split the sentence into individual words
  words = sentence.split()

  # Use a list comprehension to remove stop words
  filtered_words = [word for word in words if word not in stop_words]

  # Join the filtered words back into a sentence
  return ' '.join(filtered_words)

def process_entity(s):
    """All-purpose text cleaning of very dirty entities list"""

    _s = unidecode(s).lower()

    if any([bad in _s for bad in ["http", "www.", "t.me", ">>", "@", "/", ">"]]):
        return None
    if _s.startswith("@"):
        return None
    # if len(emoji.emoji_list(_s)) > 0:
    #     return None
    if len(re.findall("[\U0001F1E6-\U0001F1FF]", _s)) > 0:
        return None
    if len(re.findall(r"^[^a-zA-Z0-9]$", _s)) == 1:
        return None

    _s = re.sub("\s\s+", " ", _s)

    if _s.endswith("'s"):
        _s = "'s".join(_s.split("'s")[:-1])
    if _s.endswith("’s"):
        _s = "'s".join(_s.split("’s")[:-1])
    if _s.endswith("'"):
        _s = "'".join(_s.split("'")[:-1])
    if _s.startswith("the"):
        _s = "the".join(_s.split("the")[1:])
    if _s.startswith("now - "):
        _s = "now - ".join(_s.split("now - ")[1:])
    if _s.startswith("new - "):
        _s = "new - ".join(_s.split("new - ")[1:])
    if _s.startswith("in - "):
        _s = "in - ".join(_s.split("in - ")[1:])
    if _s.startswith('"'):
        _s = '"'.join(_s.split('"')[1:])
    if _s.startswith("'"):
        _s = "'".join(_s.split("'")[1:])
    if _s.startswith("a "):
        _s = "a ".join(_s.split("a ")[1:])
    if _s.startswith("inbox -"):
        _s = "inbox -".join(_s.split("inbox -")[1:])
    if _s.endswith("- telegram"):
        _s = "- telegram".join(_s.split("- telegram")[:-1])
    _s = _s.replace("\xa0", " ")

    for bad in ")(…º×•#:|[]":
        _s = _s.replace(bad, "")
    for bad in ["\n"]:
        _s = _s.replace(bad, " ")
    _s = _s.replace("\u2800", " ")
    _s = _s.replace("u.s.", "us")
    _s = _s.replace("u s.", "us")
    _s = _s.replace("u.n.", "un")

    _s = _s.strip('&"\'".!-,*~+= ')

    if _s.startswith('"'):
        if _s.count('"') % 2 == 1:
            _s = _s[1:]
    if _s.startswith("'"):
        if _s.count("'") % 2 == 1:
            _s = _s[1:]
    for pattern in [r"(\d+)(a|p)m", r"(\d+)"]:
        if re.findall(pattern, _s):
            return None

    entity = unidecode(_s).strip().lower()

    # if renamed_entity := replacement_dict.get(entity):
    #     return renamed_entity
    # else:
    if len(entity) <= 2:
        return None
    else:
        return entity


def get_wiki_id(text, language="en", auto_suggest=False, debug=False):
  assert language == "en", "Only English is supported for now!"
  wikipedia.set_lang(language)

  if debug: print(text)

  try:
    page = wikipedia.page(text, auto_suggest=False)
  except wikipedia.DisambiguationError or wikipedia.PageError:
      return None
  except BaseException:
      if debug: print('Other exception')
      return None
  return page.pageid

# Methods to disambiguate.
class DisambiguateMethod(Enum):
    NONE = 0 # Just remove stop words.
    WIKI = 1 # Use Wikipedia IDs.

def disambiguate_entities(entities, debug=False, method=DisambiguateMethod.NONE):
  # normalized_texts = [remove_stop_words(e.text.lower()) for e in entities]
  normalized_texts = [process_entity(e.text.lower()) for e in entities]

  if debug: print(normalized_texts)

  if method==DisambiguateMethod.NONE:
    wiki_ids = [None for _ in entities]
  elif method==DisambiguateMethod.WIKI:
    wiki_ids = (get_wiki_id(entity.text, auto_suggest=False, debug=debug) for entity in entities)

  entities = [Entity(text=e.text,type=e.type, wiki_id=wiki_id, id=(wiki_id or text), normalized_text=text) for e, text, wiki_id in zip(entities, normalized_texts, wiki_ids)]

  found_ids = set()
  normalized_entities = []
  for entity in entities:
    if entity.id not in found_ids:
      normalized_entities.append(entity)
      found_ids.add(entity.id)
    else:
      continue

  return normalized_entities


disambiguate_entities(entities, debug=True)
```

```python
# @title Create "Database" of posts
import ast

# From raw text.
def process_text_to_entities(text):
  entities = extract_entities(text)
  # df['spacy_entities'] = entities
  entities = filter_entities(entities)
  # df['filtered_entities'] =
  entities = disambiguate_entities(entities)
  return entities

# From Qanon database entities already extracted.
def process_named_entities(entities, disambiguate_method=DisambiguateMethod.NONE):
  try:
    entities = ast.literal_eval(entities)
    entities = [Entity(text=e['text'],type=e['type']) for e in entities if isinstance(e,dict) and 'text' in e and 'type' in e] if isinstance(entities, list) else []
    entities = filter_entities(entities)
    entities = disambiguate_entities(entities, method=disambiguate_method)
  except Exception:
    entities = []
  return entities

def create_db_from_text(df):
  normalized_entities = df['content'].map(process_text_to_entities)
  df['normalized_ids'] = normalized_entities.map(lambda x: set([e.id for e in x]))
  df['normalized_entities'] = normalized_entities.map(lambda x: [e.text for e in x])
  return df


def create_db_from_entities(df, disambiguate_method=DisambiguateMethod.NONE):
  df['normalized_entities'] = df['named_entities'].map(lambda x: process_named_entities(x,disambiguate_method))
  df['normalized_ids'] = df['normalized_entities'].map(lambda x: set([e.id for e in x]))
  df['normalized_texts'] = df['normalized_entities'].map(lambda x: set([e.normalized_text for e in x]))
  return df
```

```python
# @title Exact match

def cluster_by_overlaps(df, debug=False):
  cols_to_agg = ['content','id'] #'raw_id', 'channel', 'date', 'url', 'author_id',  'hashtags', 'detected_language', 'forwards']

  df['overlap_id'] = df['normalized_ids'].map(lambda ids: str(sorted(set(ids))))
  agg_df = pd.concat([
    df.groupby('overlap_id').count()['id'].rename('count'),
    df.groupby('overlap_id')['normalized_entities'].apply(lambda x: list(x)[0]),
    df.groupby('overlap_id')['normalized_ids'].apply(lambda x: list(x)[0]),
    df.groupby('overlap_id')[cols_to_agg].aggregate(lambda x: list(x))
  ], axis=1).sort_values(by='count', ascending=False)

  return agg_df

overlap_df = cluster_by_overlaps(long_db)

# filter by num of entities > 3
filtered_overlap_df = overlap_df[overlap_df['normalized_entities'].map(lambda x: len(x)>3)]
filtered_overlap_df
```

## Cluster by co occurence

```python
from sklearn.cluster import AgglomerativeClustering

# Create matrix dist with pairwise distances between objects.
# https://stackoverflow.com/questions/47321133/sklearn-hierarchical-agglomerative-clustering-using-similarity-matrix
      #         prof1     prof2     prof3
      #  prof1     0        0.8     0.9
      #  prof2     0.8      0       0.2
      #  prof3     0.9      0.2     0

import numpy as np

def make_coo_matrix(df, debug=False):
  # Create coocurances matrix: https://stackoverflow.com/questions/53850559/construct-co-occurrence-matrix-from-a-list-of-list-in-python
  lst = list(df['normalized_ids'].map(list))
  u = (pd.get_dummies(pd.DataFrame(lst), prefix='', prefix_sep='')
       .groupby(level=0, axis=1)
       .sum())

  v = u.T.dot(u)
  v.values[(np.r_[:len(v)], ) * 2] = 0
  return v

def cluster_by_coocurrence(df=None, coo_matrix=None, debug=False):
  if coo_matrix is None:
    coo_matrix = make_coo_matrix(df)

  dist_matrix = coo_matrix.applymap(lambda x: 1.0/x if x>0 else 0)

  model = AgglomerativeClustering(affinity='precomputed', n_clusters=20, linkage='complete').fit(dist_matrix.to_numpy())
  return coo_matrix, model

def get_clusters_from_model(coo_matrix, model):
  df = pd.DataFrame({'entity':list(coo_matrix.columns),'cluster_id':model.labels_})
  cluster_df = pd.concat([
      df.groupby('cluster_id').count()['entity'].rename('count'),
      df.groupby('cluster_id')['entity'].apply(lambda x: list(x)).rename('entities'),
    ], axis=1).sort_values(by='count', ascending=False)
  return cluster_df

def get_clusters_from_gephi(nodes_df):
  cluster_df = pd.concat([
      nodes_df.groupby('cluster_id').count()['entity'].rename('count'),
      nodes_df.groupby('cluster_id')['entity'].apply(lambda x: list(x)).rename('entities'),
    ], axis=1).sort_values(by='count', ascending=False)
  return cluster_df


def find_posts_with_entities(entities, posts_df, min_overlap=1):
  matches = posts_df['normalized_ids'].map(lambda ids: len(set(ids).intersection(set(entities))) >= min_overlap)
  return posts_df[matches]

def find_posts_from_clusters(posts_df, cluster_df):
   cluster_df.entities.map(lambda x: find_posts_with_entities(x, posts_df))
```

```python
%pip install tensorflow-text
import tensorflow_hub as hub
import tensorflow as tf
import tensorflow_text as text  # Needed for loading universal-sentence-encoder-cmlm/multilingual-preprocess
import numpy as np

preprocessor = hub.KerasLayer(
    "https://tfhub.dev/google/universal-sentence-encoder-cmlm/multilingual-preprocess/2")
encoder = hub.KerasLayer("https://tfhub.dev/google/LaBSE/2")
```

```python
def normalization(embeds):
  norms = np.linalg.norm(embeds, 2, axis=1, keepdims=True)
  return embeds/norms

def extract_embeddings(texts):
  tf_texts = tf.constant(texts)

  embeds = encoder(preprocessor(tf_texts))["default"]

  # For semantic similarity tasks, apply l2 normalization to embeddings
  embeds = normalization(embeds)
  return embeds
```

```python
unique_entities = filtered_df['parsed_entities'].explode().drop_duplicates().values
```

```python
%%time
embeddings = extract_embeddings(unique_entities[:100])
```

```python
similarities = np.matmul(embeddings, np.transpose(embeddings))
```

```python
similarities
```

```python
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import pairwise_distances

# Convert similarity to distance (required for hierarchical clustering)
distance_matrix = 1 - similarities

# Perform hierarchical clustering
n_clusters = 10  # specify the number of clusters
clustering = AgglomerativeClustering(distance_threshold=0.9, n_clusters=None)
labels = clustering.fit_predict(distance_matrix)

# Print the cluster labels
print("Cluster Labels:", labels)
```

```python
pd.DataFrame(zip(unique_entities[:100], labels), columns=['name', 'cluster']).groupby('cluster').aggregate(lambda x: list(x))
```

```python
from detoxify import Detoxify

def analyze_text_with_detoxify(text):
    toxicity_model = Detoxify('original')

    # 'original' is one of the available models, you can choose others like 'original-multilingual' or 'unitox'
    # depending on your requirements

    results = toxicity_model.predict(text)

    # The 'toxicity' key in the results dictionary represents the toxicity score
    toxicity_score = results['toxicity']

    return toxicity_score

# Example usage:
text_to_analyze = "Kill the jews"
toxicity_score = analyze_text_with_detoxify(text_to_analyze)

print(f"Toxicity Score: {toxicity_score}")
```

```python
sentences = '\n\n'.join(filtered_df['content'].values).split('.')
```

```python
import tqdm
```

```python
def split_text_into_chunks(text, chunk_length):
    """
    Split a text into chunks of a given length.

    Parameters:
    - text (str): The input text.
    - chunk_length (int): The desired length of each chunk.

    Returns:
    - List of strings: Chunks of the input text.
    """
    return [text[i:i + chunk_length] for i in range(0, len(text), chunk_length)]

# Example usage:
input_text = "This is an example text that we want to split into chunks."
chunk_length = 10

result_chunks = split_text_into_chunks(input_text, chunk_length)
```

```python
scores = []
for sent in tqdm.tqdm(sentences):
  toxicity_score = analyze_text_with_detoxify(sent)
  scores.append(toxicity_score)
```
